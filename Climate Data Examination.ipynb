In this project, I will examine the temperature and precipitation data of 21 countries bordering the Mediterranean using the World Bank's Climate Change Information Portal between 1950 and 2020.

  (Spain, France, Monaco, Italy, Slovenia, Croatia, Bosnia and Herzegovina, Montenegro, Albania, Greece, Turkey, Syria, Lebanon, Israel, Egypt, Libya, Tunisia, Algeria, Morocco, Malta and Cyprus.)

Things I will do with the data set:

- Computing the average surface temperature per month for each country, saving the results in a pandas data frame called MAST, and displaying the data using a line plot.
- Computing the average precipitation per month for each country, saving the results in a pandas data frame called MAPR, and displaying the data using a line plot.
- Computing the average surface temperature per year for each country, saving the results in a pandas data frame called YAST, and displaying the data using a line plot.
- Computing the average precipitation per year for each country, saving the results in a pandas data frame called YAPR, and displaying the data using a line plot.

```python
import pandas as pd
import json
import requests
import datetime
import matplotlib.pyplot as plt
```

In the cell above, I imported "pandas" library to process the data, "requests" library to use the API link in the project, "datetime" library to be clear on dates and "json" library to process the data in json format.

```python
def process_data(response, start_date="1950-01", end_date="2019-12"):
    data = response.json()
    df = pd.DataFrame(data["data"])

    selected_data = df.loc[start_date:end_date]
    selected_data.reset_index(inplace=True)


    selected_data = selected_data.rename(columns={'index': 'Date'})

    return selected_data
```

In the table below, you see the "Average Surface Temperature" value of each month between 1950 and 2020. In this way, I have now cleared the data. I can work on it.

```python
response_temperature = requests.get("https://cckpapi.worldbank.org/cckp/v1/cru-x0.5_timeseries_tas_timeseries_monthly_1901-2022_mean_historical_cru_ts4.07_mean/GRC,ISR,TUR,ESP,REU,MCO,ITA,SVN,HRV,BIH,MNE,ALB,SYR,LBN,EGY,LBY,TUN,DZA,MAR,MLT,CYP?_format=json")
processed_data_temperature = process_data(response_temperature)

processed_data_temperature
```

In the table below, you see the "Average Precipitation" value of each month between 1950 and 2020. In this way, I have now cleared the data. I can work on it

```python
response_precipitation = requests.get("https://cckpapi.worldbank.org/cckp/v1/era5-x0.5_timeseries_pr_timeseries_monthly_1950-2020_mean_historical_era5_era5_mean/GRC,ISR,TUR,ESP,REU,MCO,ITA,SVN,HRV,BIH,MNE,ALB,SYR,LBN,EGY,LBY,TUN,DZA,MAR,MLT,CYP?_format=json")
processed_data_precipitation = process_data(response_precipitation)

processed_data_precipitation
```

I created the below function to be able to calculate the desired average values. I added the "period" variable to be able to use the function on both annual and monthly data.

And I created an if-else block. In this if-else block, annual or monthly data will be processed depending on how the function is called with "period". In this way, monthly or annual data can be calculated without mixing them together. And it groups data acording to desired date type, so we don't get confused.

I used datetime elements when making date-related adjustments.

This function returns a table of the monthly or annual average of the desired value.

```python
def calculate_averages(data, period):
    df_data = pd.DataFrame(data)
    df_data['Date'] = pd.to_datetime(df_data['Date'], format='%Y-%m')
    
    if period == 'Monthly':
        df_data['Month'] = df_data['Date'].dt.month
        result = df_data.groupby('Month').mean().reset_index()
    elif period == 'Annual':
        df_data['Year'] = df_data['Date'].dt.year
        result = df_data.groupby('Year').mean().reset_index()
    else:
        return None
    
    result = result.drop(columns=['Date'])
    
    return result
```

## Monthly Average Air Surface Temperature

```python
MAST = calculate_averages(processed_data_temperature, period = 'Monthly' )

MAST
```

## Annual Average Air Surface Temperature

```python
YAST = calculate_averages(processed_data_temperature, period = 'Annual' )

YAST
```

## Monthly Average Precipitation

```python
MAPR = calculate_averages(processed_data_precipitation, period = 'Monthly')

MAPR
```

## Annual Average Precipitation

```python
YAPR = calculate_averages(processed_data_precipitation, period = 'Annual' )

YAPR
```

The below function allows visualizing the data and tables we have obtained. Since I wanted to be able to use this function in all my graphs, I added data_name and period variables. Thanks to the data_name variable, I can use the function for both "temperature" and "precipitation" situations. The period variable makes the function suitable for monthly and yearly situations.

Finally, since the first column is the date column, I don't want to see it on the chart. So I wrote exclude_first_column=True and prevented the date variable from being plotted on the chart.

I also defined the exclude_first_column variable in the If-Else block and made it work properly.

To plot the graph, I started plotting the data I grouped by country, one by one, with a for loop.

I named the x-axis period and the y-axis Average value. To make the graph easier to understand, I wrote what the graph is in the title.

I used the plt.legend(loc='best') command to select the most suitable ranges for the values of the chart.

Finally, I completed the visualization by drawing the graph.

```python
def visualize_averages(dataframe, data_name, period, exclude_first_column=True):
    if exclude_first_column:
        columns_to_plot = dataframe.columns[1:]
    else:
        columns_to_plot = dataframe.columns

    for country in columns_to_plot:
        plt.plot(dataframe.index, dataframe[country], label=country)

    plt.xlabel(f'{period}')
    plt.ylabel(f'Average {data_name}')
    plt.title(f'{period} Average {data_name}')
    plt.legend(loc='best')

    plt.show()
```

```python
visualize_averages(MAST, "Surface Temperature", "Monthly")
```

```python
visualize_averages(MAPR, "Precipitation", "Monthly")
```

```python
visualize_averages(MAPR, "Precipitation", "Annual")
```

```python
visualize_averages(YAST, "Surface Temperature", "Annual")
```
